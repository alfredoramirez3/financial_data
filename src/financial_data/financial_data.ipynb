```python
import pandas as pd
import yfinance as yf
print(pd.__version__)
print(yf.__version__)
```

```python
sp_wiki_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
ldf = pd.read_html(sp_wiki_url)
```

```python
len(ldf)
```

```python
df = ldf[0]
df.head()
```

```python
ticker_list = list(df['Symbol'].values)
```

```python
ticker_list
```

```python
yf_df = yf.download(ticker_list)
```

```python
yf_df.shape
```

```python
yf_df.head()
```

```python
adj_close = yf_df['Adj Close']
```
